### Описание датасета:
Данный датасет представляет собой информацию о книгах, пользующихся наибольшим спросом в списке бестселлеров Amazon в разделе Книги. Данные были собраны с помощью веб-скрипта, который просматривает несколько страниц списка бестселлеров. Набор данных содержит такие сведения, как название книги, имя автора, рейтинг и цена. Данные являются реальными и взяты с открытого источника Kaggle (https://www.kaggle.com/datasets).

### Описание атрибутов датасета:
В выбранном датасете представлены два атрибута:
- **Стоимость**: представляет собой стоимость за конкретное произведение. Тип данных - числовой.
- **Рейтинг**: представляет собой значение в рейтинге произведения относительно других книг среди бестселлеров. Тип данных - числовой.

### Описание задачи анализа:
Задача анализа данного датасета заключается в изучении изменения стоимости от позиции в рейтинге бестселлеров.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('BestSeller_Books_of_Amazon.csv', delimiter=',')

mean_Rating = df['Rating'].mean()
std_Rating = df['Rating'].std()
print("Среднее значение для атрибута 'Rating': ", mean_Rating)
print("Стандартное отклонение для атрибута 'Rating': ", std_Rating)

std_Price = df['Price'].std()
mean_Price = df['Price'].mean()
print("Среднее значение для атрибута 'Price': ", mean_Price)
print("Стандартное отклонение для атрибута 'Price': ", std_Price)
```

```python
plt.hist(df['Price'], bins=50)
plt.xlabel('Стоимость')
plt.ylabel('Частота')
plt.title('Гистограмма распределения стоимости')
plt.show()

plt.hist(df['Rating'], bins=50)
plt.xlabel('Рейтинг')
plt.ylabel('Частота')
plt.title('Гистограмма распределения рейтинга')
plt.show()
```

Как видно из приведенных диаграмм, выбросов у данных атрибутов нет.

```python
# Проверка на наличие пропущенных значений
missing_values = df.isnull().sum()
print("Количество пропущенных значений:")
print(missing_values)
```

Пропущенных значений у данных атрибутов нет.

```python
# Корреляция между параметрами
correlation = df[['Rating', 'Price']].corr()

# Построение графика рассеивания
sns.scatterplot(x='Rating', y='Price', data=df)
plt.title('График рассеивания')
plt.show()

print("Корреляция между стоимостью и рейтингом:")
print(correlation)
```

Таким образом, чем ближе значение корреляции к 1 по модулю, тем сильнее связь между стоимостью и местом в списке лучших книг.

Положительная корреляция между стоимостью и рейтинговой позицией означает, что с увеличением стоимости также увеличивается рейтинг. Это может указывать на наличие общей причины, которая влияет на оба явления, хотя они не взаимодействуют напрямую.
